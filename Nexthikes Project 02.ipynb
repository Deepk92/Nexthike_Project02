### Validate the correctness of the data 

```python
# Check missing values

concatenated_data.isnull().sum()
```

```python
# Check for duplicates based on all columns

duplicate_rows = concatenated_data[concatenated_data.duplicated()]

# Remove duplicates

concatenated_data.drop_duplicates()
```

### Identify outliers (for a specific column)

```python
Q1 = concatenated_data['cnt'].quantile(0.25)
Q3 = concatenated_data['cnt'].quantile(0.75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

outliers = concatenated_data[(concatenated_data['cnt'] < lower_fence)  | (concatenated_data['cnt'] > upper_fence)]

print('Potential Outliers :')
print(outliers)
```

```python
outliers
```

```python
outliers.shape
```

```python
# Remove Outliers

df = concatenated_data[(concatenated_data['cnt'] >= lower_fence) & (concatenated_data['cnt'] <= upper_fence)]
```

```python
df
```

```python
df1 = df.reset_index()
```

```python
df1
```

```python
# Drop unnecessary columns

Final_Data = df1.drop(['index', 'instant'], axis = 1)
```
